Testing Semua Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from minisom import MiniSom
from sklearn.decomposition import TruncatedSVD
import requests

# 1. Test Pandas & Numpy
df = pd.DataFrame(np.random.randint(0,100,size=(5, 4)), columns=list('ABCD'))
print("Pandas OK - Test Dataframe:\n", df.head(2))

# 2. Test Scikit-learn (SVD)
svd = TruncatedSVD(n_components=2)
svd.fit(df)
print("\nScikit-learn OK - SVD Components:", svd.n_components)

# 3. Test MiniSom
som = MiniSom(2, 2, 4) 
print("MiniSom OK - Initialization successful")

# 4. Test Requests (Koneksi ke Wikidata API)
try:
    response = requests.get("https://www.wikidata.org/w/api.php", timeout=5)
    print("Requests OK - Connection to Wikidata: Status", response.status_code)
except:
    print("Requests Warning - No internet connection, but library is installed.")

# 5. Test Matplotlib (Visualisasi)
plt.figure(figsize=(2,2))
plt.title("Visual Test")
plt.plot([1, 2], [1, 2])
plt.show()
print("Visualisasi OK - Plot berhasil ditampilkan")
```
